```python
import numpy as np
import sympy as sp

# 벡터 정의
a = np.array([2, 3])
b = np.array([4, 1])

# 벡터의 크기를 sympy로 계산
norm_a_sym = sp.sqrt(sum([i**2 for i in a]))
norm_b_sym = sp.sqrt(sum([i**2 for i in b]))

# 내적 계산
dot_product_sym = sp.Rational(np.dot(a, b))

# 각도의 코사인 값 계산
cos_theta_sym = dot_product_sym / (norm_a_sym * norm_b_sym)

# 라디안 각도를 도로 변환
theta_sym = sp.acos(cos_theta_sym)
degree_theta_sym = sp.deg(theta_sym)

norm_a_sym, norm_b_sym, cos_theta_sym, theta_sym, degree_theta_sym
```

## 벡터의 내적을 구하기

```python
import numpy as np
import matplotlib.pyplot as plt

origin = np.array([0, 0])
vector_x = np.array([4, 0])
vector_y = np.array([0, 2])
vector_r = np.array([4, 2])

plt.quiver(*origin, *vector_r, angles="xy", scale_units="xy", scale=1, color='g', label='Vector a')
plt.quiver(*origin, *vector_x, angles="xy", scale_units="xy", scale=1, color='r', label='Vector b')

plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')

print("DotProduct  : ", np.dot(vector_x,vector_r))
print(np.degrees(np.arccos(np.dot(vector_x,vector_r) / (np.linalg.norm(vector_x) * np.linalg.norm(vector_r)))))

plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the origin and vectors
origin = np.array([0, 0])
vector_x = np.array([4, 0])
vector_y = np.array([0, 2])
vector_r = np.array([4, 2])

# Plot the vectors
plt.quiver(*origin, *vector_r, angles="xy", scale_units="xy", scale=1, color='g', label='Vector a')
plt.quiver(*origin, *vector_x, angles="xy", scale_units="xy", scale=1, color='r', label='Vector b')

# Set plot limits
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid()
plt.gca().set_aspect('equal', adjustable='box')

# Calculate dot product
dot_product = np.dot(vector_x, vector_r)
print("Dot Product:", dot_product)

# Calculate the angle between the vectors
angle_radians = np.arccos(dot_product / (np.linalg.norm(vector_x) * np.linalg.norm(vector_r)))
angle_degrees = np.degrees(angle_radians)
print("Angle (radians):", angle_radians)
print("Angle (degrees):", angle_degrees)

# Annotate the angle on the plot
angle_text = f'{angle_degrees:.2f}°'
plt.text(2, 1, angle_text, fontsize=12, bbox=dict(facecolor='white', alpha=0.6))

# Show legend and plot
plt.legend()
plt.show()
```

# 직교 벡터가 의 내적은 0이라는 사실을 증명

```python
import numpy as np
import matplotlib.pyplot as plt

vector_x = np.array([0,5])
vector_y = np.array([5,0])

x_norm = np.linalg.norm(vector_x)
y_norm = np.linalg.norm(vector_y)
origin = np.array([0,0])

plt.quiver(*origin,*vector_x,angles="xy", scale_units="xy", scale=1, color='g', label='Vector a')
plt.quiver(*origin,*vector_y,angles="xy", scale_units="xy", scale=1, color='g', label='Vector b')

plt.xlim(-1,6)
plt.ylim(-1,6)
plt.grid()

dot_product = np.dot(vector_x,vector_y)
angle_radians = np.arccos(dot_product/x_norm*y_norm)
print("angle : " ,np.degrees(angle_radians))
print(dot_product)

plt.show()
```

# 비단위 벡터와 같은 방향의 단위벡터 만들기

```python
import numpy as np
import matplotlib.pyplot as plt

# 원점 좌표
origin = np.array([0, 0])

# 원래 벡터
vector = np.array([3, 5])

# 단위 벡터 계산
vector_unit = (1 / np.linalg.norm(vector)) * vector

print(np.linalg.norm(vector_unit))

# 벡터와 단위 벡터 시각화
plt.quiver(*origin, *vector, angles="xy", scale_units="xy", scale=1, color='g', label='Vector a')
plt.quiver(*origin, *vector_unit, angles="xy", scale_units="xy", scale=1, color='r', label='Unit Vector')

# 그래프 설정
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.grid()
plt.legend()
plt.show()
```
