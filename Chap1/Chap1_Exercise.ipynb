# 연습문제 1 - 3
## 벡터를 입력으로 받아 동일한 방향의 단위 벡터를 출력하는 함수를 구현
### 영벡터를 입력하면 어떻게 될까?

```python
import numpy as np
import matplotlib.pyplot as plt

# 원점 좌표
origin = np.array([0, 0])

# 원래 벡터
vector = np.array([3, 5])

# 단위 벡터 계산
vector_unit = (1 / np.linalg.norm(vector)) * vector

print(np.linalg.norm(vector_unit))

# 벡터와 단위 벡터 시각화
plt.quiver(*origin, *vector, angles="xy", scale_units="xy", scale=1, color='g', label='Vector a')
plt.quiver(*origin, *vector_unit, angles="xy", scale_units="xy", scale=1, color='r', label='Unit Vector')

# 그래프 설정
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.grid()
plt.legend()
plt.show()
```

# 연습문제 1-4
## 벡터와 원하는 크기를 입력 받고, 벡터와 동일한 방향이지만 두번 째 입력에 해당하는 크기를 가진 벡터를 반환하자

```python
import numpy as np
import matplotlib.pyplot as plt

INPUT = 3
origin = np.array([0,0])
vector = np.array([1,4])

vector_unit = ((1 / np.linalg.norm(vector)) * vector) * INPUT

plt.quiver(*origin,*vector,angles="xy",scale_units="xy",scale=1,color="g", label="vector a")
plt.quiver(*origin,*vector_unit,angles="xy",scale_units="xy",scale=1,color="r", label="vector a")
plt.grid()
plt.xlim(-1,6)
plt.ylim(-1,6)
plt.show()
print(np.linalg.norm(vector_unit))
```
